```python
import pandas as pd
import numpy as np
import calendar
```

```python
fn = 'pittsburgh2013_weather.csv'
```

```python
pittsburghWeather = pd.read_csv(fn, parse_dates= ['Date'])
print(pittsburghWeather.info())
pittsburghWeather.head()
```

```python
pittsburghWeather['month'] = pittsburghWeather.Date.dt.month
cols = pittsburghWeather.columns
newCols = [cols[0]]+[cols[-1]]+cols[1:-1].tolist()
pittsburghWeather = pittsburghWeather[newCols]
out = pittsburghWeather.groupby('month')['Max TemperatureF'].max().reset_index()
out['month']=out['month'].apply(lambda x: calendar.month_abbr[x], convert_dtype = False)
out
```

```python
out.sort_values(by = 'Max TemperatureF', inplace = True, ascending = False)
out
```

*We'll now convert the data to a time series, and do some quarterly resampling (sampling at the <b>beginning</b> of the quarter. We'll resample the 'Max TemperatureF' column, aggregating using the max function and resample the 'Mean TemperatureF' aggregating using the mean function.*

**See this link for more info: http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases**

```python
pittsburghWeather.set_index('Date', inplace = True)
quarterlyResampled_maxes = pittsburghWeather.loc[:,'Max TemperatureF'].resample('QS').max().to_frame().reset_index() 
#NOTE: in above, 'QS' is 'quarterly start'; this is start of the quarter (i.e. Jan, Apr, July, etc.). Resampling w/ 'Q' would
#be end of quarter (i.e. March, June, etc.)
quarterlyResampled_maxes['Date']=quarterlyResampled_maxes['Date'].dt.month.apply(lambda x: calendar.month_abbr[x], convert_dtype = False)
quarterlyResampled_maxes = quarterlyResampled_maxes.set_index('Date')
quarterlyResampled_maxes
```

```python
quarterlyResampled_means = pittsburghWeather.loc[:,'Mean TemperatureF'].resample('QS').mean().to_frame().reset_index()
quarterlyResampled_means['Date']=quarterlyResampled_means['Date'].dt.month.apply(lambda x:calendar.month_abbr[x], convert_dtype = False)
quarterlyResampled_means = quarterlyResampled_means.sort_values(by = 'Date').set_index('Date')
quarterlyResampled_means
```

*We'll now reindex the quarterlyResampled_means dataframe (above) to include all months and this will  yield NA values*

*See calendar  module documentation for more information on calendar.month_abbrev used below: https://docs.python.org/3/library/calendar.html*

```python
months = list(calendar.month_abbr)[1:]
```

```python
quarterlyResampled_means.reindex(months)
```

*To fill in the NaNs we'll have to use the ffill method so that November and December aren't left with NaNs- as they would be if bfill were used*

```python
quarterlyResampled_means.reindex(months).ffill()
```
